```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
fifa_data = "Resources/fifa_ranking.csv"
fifa_df = pd.read_csv(fifa_data)

fifa_df.head()
```

```python
##Economic Freedom data examines information from 2016. Since FIFA date ranges from 1993-2018, the rank_date format is changed to datetime.
```

```python
fifa_df['date'] = pd.to_datetime(fifa_df['rank_date'])
fifa_df.head()
```

```python
country_data = "Resources/efw_cc.csv"
country_df = pd.read_csv(country_data)
country_df.head()
```

```python
##Fifa Data Filtered by Country, Rank, Ctry_ID and Date
```

```python
fifa_cols = [ "country_full", "rank", "country_abrv", "date"]
fifa_transformed = fifa_df[fifa_cols].copy()

fifa_transformed = fifa_transformed.rename(columns={"country_full": "country", "country_abrv": "ctry_id"})
fifa_transformed.set_index("country", inplace=True)

fifa_transformed.head()
```

```python
##FIFA Date filtered to Team Ranked at the end of 2016 (last ranking occurred on 12/22/16)
```

```python
fifa_filtered = fifa_transformed.loc[fifa_transformed['date'] == '2016-12-22']
fifa_filtered.head()
```

```python
##Duplicates removed from the Economic Freedom data
```

```python
country_cols = ["countries", "ECONOMIC FREEDOM", "rank", "year"]
country_transformed = country_df[country_cols].copy()

country_transformed = country_transformed.rename(columns={"countries": "country", "ECONOMIC FREEDOM": "freedom_score"})
country_transformed = country_transformed.drop_duplicates('country')
country_transformed.set_index("country", inplace=True)

country_transformed.head()
```

```python
connection_string = "postgres:postgres@localhost:5432/fifaEFW_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
fifa_filtered.to_sql(name='fifa_ranking', con=engine, if_exists='append', index=True)
```

```python
country_transformed.to_sql(name='econ_freedom', con=engine, if_exists='append', index=True)
```

```python
pd.read_sql_query("select * from fifa_ranking", con=engine).head()
```

```python
pd.read_sql_query("select * from econ_freedom", con=engine).head()
```
